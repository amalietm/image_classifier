```python
# importing libraries
import tensorflow as tf
import tensorflow.keras as keras
import numpy as np
```

```python
# importing drive to access pictures
from google.colab import drive
drive.mount("/content/drive")
```

```python
# load model 
model = keras.models.load_model("/content/drive/MyDrive/image_classifier/")
```

```python
# load data
images_path = "/content/drive/MyDrive/image_classifier/image_classifier_data/exam_images"
dataset = keras.preprocessing.image_dataset_from_directory(images_path)
class_names= dataset.class_names
print(class_names)
```

```python
# use model to predict classification on image not included in training or validation sets
# image input
img = tf.keras.utils.load_img("/content/drive/MyDrive/image_classifier/image_classifier_data/test/deer1.jpeg", target_size=(150,150))

# convert image to array
img_array = tf.keras.utils.img_to_array(img)
# adds fourth dimension
img_array = tf.expand_dims(img_array, 0)

# make prediction
predictions = model.predict(img_array)
score = tf.nn.softmax(predictions[0])

# prints predicted class of image input
print(
    "This image most likely belongs to {}"
    .format(class_names[np.argmax(score)])
    )
```
